```python
import seaborn as sns

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\User\\Desktop\\Machine Learning and Data Science\\UNZIP_FOR_NOTEBOOKS_FINAL\DATA\\dm_office_sales.csv")

# Rugplot
plt.figure(figsize=(5,7), dpi=300)
sns.rugplot(x='salary', data=df, height=0.2)

# displot
# style options: white, dark, whitegrid, darkgrid, ticks
sns.set(style='darkgrid')
sns.displot(x='salary', data=df, bins=20, color='blue', edgecolor='red', linewidth=2, kde=True, rug=True)

# plotting only kde plots
sns.kdeplot(x='salary', data=df)
sns.kdeplot(x='salary', data=df, clip=[0,160000], bw_adjust=0.1, fill=True)
```
