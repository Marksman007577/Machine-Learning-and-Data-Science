```python
# Groupby function helps to examine data on a per category basis

import numpy as np
import pandas as pd

df = pd.read_csv("C:\\Users\\User\\Desktop\\Machine Learning and Data Science\\UNZIP_FOR_NOTEBOOKS_FINAL\\DATA\\mpg.csv")

# To decide what column to use for analysis

df['model_year'].unique()
#df['name'].unique()

# If we decide to analyse using model year. 
# Now we want to find the mean of the miles per gallon for each model year alone
df.groupby('model_year').mean()['mpg']

# Now we want to find the mean of the other data after we have model year as our base for analysis
df.groupby('model_year').mean()

# We might want to find the mean for x-numbers of base parameter for analysis. e.g we take model year and cylinders
df.groupby(['model_year','cylinders']).mean()
```
