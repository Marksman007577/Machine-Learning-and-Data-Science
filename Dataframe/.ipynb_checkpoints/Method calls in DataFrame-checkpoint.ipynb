```python
# Method calls in dataframe
import numpy as np
import pandas as pd

df = pd.read_csv("C:\\Users\\User\\Desktop\\Machine Learning and Data Science\\UNZIP_FOR_NOTEBOOKS_FINAL\DATA\\tips.csv")

#1.1 Apply method on a single column
# Usecase: using the data in the total_bill column to creating a rating from one to five star based on how wxpwnsive it is

def big_mama_kitchen(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"

df['big_mama_kitchen'] = df['total_bill'].apply(big_mama_kitchen)

#1.2 Apply method on a multiple column

def tip_quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Others"


df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df:tip_quality(df['total_bill'],df['tip']),axis=1)

#2.1 Describing method
des = df.describe()
trans_des = df.describe().transpose()

#2.2 Sorting method
# usecase1: sorting by a single column
low_to_high = df.sort_values('tip') # sorts from lowest to highest
high_to_low = df.sort_values('tip', ascending=False) # sorts from highest to lowest 

#usecase 2: sorting multiple column
mul_sort = df.sort_values(['tip', 'size']) # sorts from lowest to highest

# usecase3: sorting maximum and minimum values and their index location
#min_val = df.['total_bill'].min()
#min_val_index = df.['total_bill'].idxmin()

#max_val = df.['total_bill'].max()
#max_val_index = df.['total_bill'].idxmax()

# Getting correlation values in different columns
#N.B: Only works with number data

corr_val = df.corr()

# Value Count Method. 
# Used for categorical columns e.g male vs female
cat_count = df['sex'].count()

# Unique and nunique  method
# Returns  the unique data in a data
unique_data = df['day'].unique() # returns the unique data in the dataset
nunique_data = df['day'].nunique() # returns the number of unique data in the dataset

# Replace method 
# Usecase: replacing  a column title with an acronym
# Recommended for few items to be replaced
replaced_acronym = df['sex'].replace('Female', 'F') # for single data
replaced_acronym_mul = df['sex'].replace(['Female','Male'], ['F','M']) # for multiple data

# Mapping Method
# Allows for processing and transforming all the items in an iterable(list), without using any explicit or loop
# Recommended for replacing large data
my_map = {'Female':'F', 'Male':'M'}
my_map_data = df['sex'].map(my_map)

# Finding duplicates
duplicate_data = df.duplicated() # gives true or false on the firt instance of a duplicate
duplicate_data_drop = df.drop_duplicates() # gives true or false on the firt instance of a duplicate

# Between Method
# Tells the values between your chosen values

num_between_values = df[df['total_bill'].between(5,12,inclusive=True)] #inclusive will consider your search limits of 10 and 20

# finding largest number of rows or columns with n-conditions
n_largest_tip = df.nlargest(10, 'tip')
n_smallest_tip = df.nsmallest(2, 'tip')

# Sampling method
# samples random rows within a data set or data frame

sampled_data = df.sample(5) # samples randomly any 5 columns
sampled_data_frac = df.sample(frac=0.5)
```
